```python
import numpy as np
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
```

```python
ar1 = ArmaProcess(ar=np.array([1., -0.8]), ma=None, nobs=100)
ar1_samples = ar1.generate_sample()
plt.plot(ar1_samples)
plt.show()
```

```python
plt.scatter(
    ar1_samples,
    2. + 1 * ar1_samples - 0.3 * ar1_samples**2 - 0.1 * ar1_samples**3
)
plt.show()
```
